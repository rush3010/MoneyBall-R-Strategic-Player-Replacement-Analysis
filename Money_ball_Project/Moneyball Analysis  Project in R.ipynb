```r
#batting average(BA)
#BA = H/AB
df_bat$BA = df_bat$H / df_bat$AB
```

```r
head(df_bat$BA)
```

```r
#on base percentage(OBP)
#obp = (H+bb+hbp)/(ab+bb+hbp+sf)
df_bat$OBP = (df_bat$H + df_bat$BB + df_bat$HBP)/(df_bat$AB + df_bat$BB + df_bat$HBP + df_bat$SF)
```

```r
head(df_bat$OBP)
```

```r
tail(df_bat$OBP)
```

```r
#singles(X1B)
#singles = h-2b-3b-hr
df_bat$X1B = (df_bat$H - df_bat$X2B - df_bat$X3B - df_bat$HR)
```

```r
head(df_bat$X1B)
```

```r
#slugging percentage(SLG)
# slg=(x1b+2*x2b+3*x3b+4*hr)/ab
df_bat$SLG = (df_bat$X1B+2*df_bat$X2B+3*df_bat$X3B+4*df_bat$HR)/df_bat$AB
```

```r
head(df_bat$SLG)
```

```r
#On Base PLus Slugging (OPS)
#OPS determines combine how well a hitter can reach base, with how well he can hit for average and for power.
#OPS = OBP + SLG
df_bat$OPS = df_bat$OBP + df_bat$SLG
```

# Finding new players to replace

Few conditions of replacement : 
1. The total combined salary of the three players can not exceed 15 million dollars.
2. Their combined number of At Bats (AB) needs to be equal to or greater than the lost players.
3. Their mean OBP had to equal to or greater than the mean OBP of the lost players

```r
repl_players=subset(newdf,yearID == 2001)
```

```r
repl_players
```

```r
#checking mean OBP of lost players
meanOBP_lost_players=mean(lost_players$OBP)
meanOBP_lost_players
```

```r
#checking combined AB of lost players
total_AB_lost_players=sum(lost_players$AB)
total_AB_lost_players
```

```r
repl_players[is.na(repl_players)]=0
repl_players
```

```r
repl_players = filter(repl_players,OBP>0)
repl_players
```

```r
#checking ops above 0.800
a=filter(repl_players,OPS>=0.800)
a %>% arrange(desc(OPS))
```

```r
x=a[,c('playerID','teamID.x','lgID.x','AB','H','X2B','X3B','HR','BA','OBP','SLG','salary','OPS')]
b=x %>% arrange(desc(OPS)) %>% filter(.,salary<7500000) %>% filter(.,AB>100) 
y=head(b,20)
```

#### Considering above table, based on On Base plus Slugging(OPS) score possible replacement for lost players are :

```r
replacement=subset(repl_players,playerID %in% c('gonzalu01','berkmla01','heltoto01'))
replacement
```

```r
#checking mean OBP of replacement players
meanOBP_new_players=mean(replacement$OBP)
cat("Mean OBP of repalacement players : ",meanOBP_new_players)
cat("\n Mean OBP of lost players : ",meanOBP_lost_players)
#Clearly, mean OBP of replacement players is better than lost players
```

```r
#checking Combined AB of replacement players
total_AB_new_players=sum(replacement$AB)
cat("Combined AB of repalacement players : ",total_AB_new_players)
cat("\n Combined AB of lost players : ",total_AB_lost_players)
#Clearly, Combined AB of replacement players is more than lost players
```
